```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.

all_returns= pd.concat([whale_pd, algo_ran, sp500returns_bang], axis="columns", join="inner")
all_returns.head()
```

### Conduct Quantitative Analysis

```python
 # Plot daily returns of all portfolios
all_returns.plot(figsize=(15,8))
```

### Calculate and Plot cumulative returns.

```python
# Calculate cumulative returns of all portfolios
# Plot cumulative returns

cumulative_returns_boom = (1+all_returns).cumprod() -1
cumulative_returns_boom.plot(figsize=(15,8))
```

### Risk Analysis

```python
# Box plot to visually show risk

all_returns.plot.box(figsize=(20,8))
```

### Calculate Standard Deviations

```python
# Calculate the daily standard deviations of all portfolios

daily_std = all_returns.std()
print(daily_std)
```

###  Determine which portfolios are riskier than the S&P 500

```python
# Calculate  the daily standard deviation of S&P 500
# Determine which portfolios are riskier than the S&P 500

riskier = daily_std[daily_std > daily_std["sp500boom"]]
print(riskier) # Below Portfolios are Riskier because their std is GREATER than SP500
```

### Calculate the Annualized Standard Deviation

```python
annualized_std = daily_std * np.sqrt(252)
annualized_std
```

### Rolling Statistics

### Calculate and plot rolling std for all portfolios with 21-day window

```python
all_returns.rolling(window=21).std().plot(figsize=(15,8))
```

###  Calculate and plot the correlation

```python
import seaborn as sns
correlation = all_returns.corr()
sns.heatmap(correlation, vmin=-1, vmax=1,annot=True,)
correlation
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
# Calculate variance of S&P 500

covariance = all_returns['BERKSHIRE HATHAWAY INC'].cov(all_returns['sp500boom'])
variance = all_returns['sp500boom'].var()
beta = covariance / variance
beta
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Computing beta
# Plot beta trend
rolling_covariance = all_returns['BERKSHIRE HATHAWAY INC'].rolling(window=60).cov(all_returns['sp500boom'])
rolling_variance = all_returns['sp500boom'].rolling(window=60).var()
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='Rolling 30-Day Beta of Berkshire Hathaway')
```
